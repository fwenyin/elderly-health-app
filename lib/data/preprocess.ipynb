```python
import pandas as pd
import ast
import re
import nltk
import csv
nltk.download('stopwords')
nltk.download('omw-1.4')
nltk.download('punkt')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
file_path = 'raw_medical_kb.csv' 
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
def clean_and_format_string(s):
    s = s.replace('[', '').replace(']', '').replace("'", "")
    s = s.replace(',', ' ')
    return s

df['symptom'] = df['symptom'].apply(clean_and_format_string)
```

```python
# df['diagnosis'] = 'Cause:\n' + df['cause'] + '\n\nPrevention:\n' + df['prevent']
```

```python
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    words = word_tokenize(text)
    cleaned_text = ' '.join([lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')])
    return cleaned_text

df['symptom'] = df['symptom'].apply(preprocess_text)
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
output_file_path = 'preprocessed_medical_kb.csv'  
df.to_csv(output_file_path, index=False, quoting=csv.QUOTE_NONNUMERIC)
```
